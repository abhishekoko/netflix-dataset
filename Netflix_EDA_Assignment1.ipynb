```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
dataset_url = 'netflix_customer_churn.csv'
df = pd.read_csv(dataset_url)
```

## Task 1: Data Overview

```python
# top 5
df.head()
```

## Task 2: Univariate Analysis

```python
# Distribution plot for 'age'
plt.figure(figsize=(10,6))
sns.histplot(df['age'],bins=20)
plt.title('Age Distribution')
plt.show()
```

```python
# Distribution plot for 'watch_hours'
plt.figure(figsize=(10,6))
sns.histplot(df['watch_hours'], kde=True, bins=20)
plt.title('Watch Hours Distribution')
plt.show()
```

```python
# Distribution plot for 'monthly_fee'
plt.figure(figsize=(10,6))
sns.histplot(df['monthly_fee'], kde=True, bins=20)
plt.title('Monthly Fee Distribution')
plt.show()
```

```python
# Distribution plot for 'churned'
plt.figure(figsize=(10,6))
sns.countplot(x='churned', data=df)
plt.title('Churned Distribution (0 = No, 1 = Yes)')
plt.show()
```

```python
# Count plot for categorical variables
plt.figure(figsize=(10,6))
sns.countplot(x='subscription_type', data=df)
plt.title('Subscription Type Distribution')
plt.show()
```

```python
# Count plot for 'gender'
plt.figure(figsize=(10,6))
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
# Count plot for 'region'
plt.figure(figsize=(10,6))
sns.countplot(x='region', data=df)
plt.title('Region Distribution')
plt.show()
```

```python
# Count plot for 'device'
plt.figure(figsize=(10,6))
sns.countplot(x='device', data=df)
plt.title('Device Distribution')
plt.show()
```

```python
# Count plot for 'payment_method'
plt.figure(figsize=(10,6))
sns.countplot(x='payment_method', data=df)
plt.title('Payment Method Distribution')
plt.show()
```

```python
# Count plot for 'favorite_genre'
plt.figure(figsize=(10,6))
sns.countplot(x='favorite_genre', data=df)
plt.title('Favorite Genre Distribution')
plt.show()
```

## Task 3: Bivariate Analysis

```python
# Compare average 'watch_hours' and 'monthly_fee' across subscription_type
plt.figure(figsize=(10,6))
sns.barplot(x='subscription_type', y='watch_hours', data=df)
plt.title('Average Watch Hours by Subscription Type')
plt.show()
```

```python
# Compare 'monthly_fee' across subscription_type
plt.figure(figsize=(10,6))
sns.barplot(x='subscription_type', y='monthly_fee', data=df)
plt.title('Average Monthly Fee by Subscription Type')
plt.show()
```

```python
# Compare 'watch_hours' and 'monthly_fee' across region
plt.figure(figsize=(10,6))
sns.barplot(x='region', y='watch_hours', data=df)
plt.title('Average Watch Hours by Region')
plt.show()
```

```python
# Compare 'watch_hours' across device type
plt.figure(figsize=(10,6))
sns.barplot(x='device', y='watch_hours', data=df)
plt.title('Average Watch Hours by Device Type')
plt.show()
```

```python
# Plot average 'avg_watch_time_per_day' by favorite_genre
plt.figure(figsize=(10,6))
sns.barplot(x='favorite_genre', y='avg_watch_time_per_day', data=df)
plt.title('Average Watch Time per Day by Favorite Genre')
plt.show()
```

```python
# Analyze churn rate by gender
plt.figure(figsize=(10,6))
sns.barplot(x='gender', y='churned', data=df)
plt.title('Churn Rate by Gender')
plt.show()
```

```python
# Analyze churn rate by region
plt.figure(figsize=(10,6))
sns.barplot(x='region', y='churned', data=df)
plt.title('Churn Rate by Region')
plt.show()
```

```python
# Analyze churn rate by subscription_type
plt.figure(figsize=(10,6))
sns.barplot(x='subscription_type', y='churned', data=df)
plt.title('Churn Rate by Subscription Type')
plt.show()
```

```python
# Analyze churn rate by payment_method
plt.figure(figsize=(10,6))
sns.barplot(x='payment_method', y='churned', data=df)
plt.title('Churn Rate by Payment Method')
plt.show()
```

## Task 4: Correlation Analysis

```python
# Create a heatmap of correlation between numerical variables
corr = df[['age', 'watch_hours', 'last_login_days', 'monthly_fee', 'number_of_profiles', 'avg_watch_time_per_day', 'churned']].corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
